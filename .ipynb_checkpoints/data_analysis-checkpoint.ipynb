```python
import pandas as pd
import re
import unicodedata
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.text import Tokenizer

%matplotlib inline
```

```python
%ls
```

```python
df = pd.read_csv("all_data.csv")
df = df.drop("Unnamed: 4", axis=1)
df.head()
```

```python
print(df["date"].unique())
```

```python
#df['numbers'] = df['document'].apply(lambda x: re.sub("[0-9]+", '', x))
def find_telefon_numbers(x):
    numbers = []
    result = re.findall("[0-9]+", x)
    for n in result:
        if (len(n)> 5):
            numbers.append(n)
    return numbers
         

def remove_accent(text):
    output = unicodedata.normalize('NFD', text).encode('ascii', 'ignore')
    decoded = output.decode('utf-8')
    return decoded

def remove_punctuatios(text):
    output = text.replace("")

df['numbers'] = df['document'].apply(lambda x: find_telefon_numbers(x))
df['lower_case'] = df['document'].str.lower()
df['lower_case'] = df['lower_case'].apply(lambda x: remove_accent(x))
df = df.sample(frac=1).reset_index(drop=True)

#lambda x: True if x % 2 == 0 else False

df.head()
```

```python
#df_demo = df.head(n=20)
#df_demo.to_json("demo_json.json")
#df_demo.to_csv("demo_data.csv", index = None, header=True)
```

```python
tokenizer = Tokenizer()
```

```python
import itertools

numbers = df["numbers"].tolist()
numbers = list(itertools.chain.from_iterable(numbers))

print(len(list(numbers)))
print(numbers[:5])
print(len(set(numbers)))
```

```python
tokenizer.fit_on_texts(numbers)

total_words = len(tokenizer.word_index) + 1
print(total_words)
```

```python
word_counts = dict(tokenizer.word_counts)
print(word_counts)
```

```python
word_counts["5212310"]
```

```python
df_counts = pd.DataFrame({'Value': list(word_counts.values()), 'Label': list(word_counts.keys()),
                         'x': np.arange(len(word_counts))}, index = list(word_counts.keys()) )


df_counts = df_counts.sort_values(['Value'], ascending=False)
df_counts.head(n=10)
```

```python
def plot_word_count(df, title):
    plt.figure(figsize=(10,6))
    ax = sns.barplot(x="x", y="Value", data=df, order=df['x'])
    ax.set_xlabel(title)
    ax.set_ylabel('Number of Occurrences')
    ax.set_xticklabels(df['Label'], rotation='vertical', fontsize=10)
    
```

```python
top20_sorted = df_counts.iloc[0:40]
top20 = df_counts.iloc[0:40]

plot_word_count(top20_sorted, f"Top {len(top20_sorted)} Words")
```

```python
plt.figure(figsize=(16, 6))

sns.boxplot(x=df_counts['Value'])
```

```python
x_o = df_counts.loc[(df_counts["Value"] < 50) & (df_counts["Value"] > 1 )]["Value"].tolist()

plt.figure(figsize=(16, 6))
sns.distplot(x_o, rug=True,norm_hist=False, kde = True)
plt.show()
```

```python
mean = np.mean(df_counts["Value"])
std = np.std(df_counts["Value"])
print(mean)
print(std)
```
