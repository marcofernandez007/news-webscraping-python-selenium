```python
import pandas as pd
import re
import unicodedata
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import timedelta, datetime

from tensorflow.keras.preprocessing.text import Tokenizer

%matplotlib inline
```

```python
df = pd.read_csv("clean_data/all_data.csv")
df = df.drop("Unnamed: 4", axis=1)
df.head()
```

```python
#df['numbers'] = df['document'].apply(lambda x: re.sub("[0-9]+", '', x))
def find_telefon_numbers(x):
    numbers = []
    result = re.findall("[0-9]+", x)
    for n in result:
        if (len(n)> 5):
            numbers.append(n)
    return numbers
         

def remove_accent(text):
    output = unicodedata.normalize('NFD', text).encode('ascii', 'ignore')
    decoded = output.decode('utf-8')
    return decoded

def remove_punctuatios(text):
    output = text.replace("")

df['numbers'] = df['document'].apply(lambda x: find_telefon_numbers(x))
df['lower_case'] = df['document'].str.lower()
df['lower_case'] = df['lower_case'].apply(lambda x: remove_accent(x))
df = df.sample(frac=1).reset_index(drop=True)

#lambda x: True if x % 2 == 0 else False

df.head()
```

```python
class NumberAnalysis:
    
    def __init__(self, number):
        
        self.number = number
        self.df_new = None
        self.ninety_five_quant = None
        self.ninety_ninth_quant = None
        self.grouped  = None
        
    def _find_outlayer(self, numbers, target):
        status = False
        for x in numbers:
            if x == target:
                status=True
                break
        return status
    
    def _extract_list(my_list):
        value = ""
        for x in my_list:
            value = x
            break
        return value
    
    def _parse_date(self, my_date):
        date_splited = my_date.split("-")
        month = date_splited[1]
        if (month[0] == "0"):
            month = month[-1]
        else:
            pass
        new_date = [month,
                    date_splited[2],
                    date_splited[0], 
                    ]
        fixed_date = '-'.join(new_date)
        print(fixed_date)
        #return datetime.strptime(fixed_date, '%m-%d-%Y')

        year = int(date_splited[0])
        month = int(month)
        day = int(date_splited[2])
        if (fixed_date == "2-31-2019"):
            day = day -  3 
            print(f"new day {day}")
        elif (fixed_date == "9-31-2018"):
            day = day - 2
        elif (fixed_date == "2-29-2018"):
            day = day - 1
        elif (fixed_date == "2-29-2019"):
            day = day- 1
        elif (fixed_date == "2-30-2018"):
            day = day -3
        elif fixed_date == "11-31-2018":
            day = day -1
        elif fixed_date == "4-31-2019":
            day -=1
        elif fixed_date == "4-31-2018":
            day -=1
        elif fixed_date == "2-30-2019":
            day -=3
        return datetime(year,month ,day)
    
    def create_target_df(self, df, number):
        self.df_new = df.loc[df['numbers'].apply(lambda x: self._find_outlayer(x,  number)) ]
    
    def extract_number_from_list(self):
        """
        Rewrite numbers column with number string instead of list of string
        input: df, standar dataframe
        output: df, input dataframe with processed numbers column.
        """
        self.df_new["numbers"] = self.df_new["numbers"].apply(lambda x: self._extract_list(x))

    def create_datestamp_column(self):
        """
        Create datestamp column
        input: df, standar dataframe
        output: df, input dataframe with new column
        """
        self.df_new["datestamp"] = self.df_new["date"].apply(lambda x: self._parse_date(x))
        
    def group_by(self):
        """
        Group by datetime column
        
        input: df, dataframe 
        output: df, grouped dataframe
        """
        self.grouped = self.df_new.groupby('datestamp').count()
        
    def calculate_quartiles(self, group_df):
        self.ninety_five_quant = group_df.quantile(.95)[0] # [0] since we only need the quantile value
        self.ninety_ninth_quant = group_df.quantile(.99)[0] # [0] since we only need the quantile value
        self.ninety_eight_quant = group_df.quantile(.98)[0] # [0] since we only need the quantile valueninety_five_quant = grouped.quantile(.95)[0] # [0] since we only need the quantile value
    
    def plot_scatter(self, group_df,):
        count = self.df_new["date"].size

        total_req =  'Total request: ' + str(count)
        nine_five_quant_str =  '95th Quantile: ' + str(self.ninety_five_quant)
        nine_eight_quant_str =  '98th Quantile: ' + str(self.ninety_eight_quant)
        nine_nine_quant_str = '99th Quantile: ' + str(self.ninety_ninth_quant)

        # x and y axis values are extracted from the grouped DataFrame
        x = self.grouped.index
        y = self.grouped.values

        print ('Plotting Graph..')
        fig = plt.figure(figsize=(16, 6))
        fig.suptitle('Scatter Plot', fontsize=14, fontweight='bold')
        ax = fig.add_subplot(111)
        fig.subplots_adjust(top=0.95)

        ax.set_xlabel('Request Time')
        ax.set_ylabel('Request Count')

        text_x_axis_value = 0.85
        ax.text(text_x_axis_value, 0.96, total_req, horizontalalignment='center', verticalalignment='center', transform = ax.transAxes)
        ax.text(text_x_axis_value, 0.88, nine_five_quant_str, horizontalalignment='center', verticalalignment='center', transform = ax.transAxes)
        ax.text(text_x_axis_value, 0.80, nine_eight_quant_str, horizontalalignment='center', verticalalignment='center', transform = ax.transAxes)
        ax.text(text_x_axis_value, 0.74, nine_nine_quant_str, horizontalalignment='center', verticalalignment='center', transform = ax.transAxes)
        ax.plot_date(x, y, xdate=True, ydate=False, color='skyblue')
        plt.show()   
```

```python
def parse_date(my_date):
    date_splited = my_date.split("-")
    month = date_splited[1]
    if (month[0] == "0"):
        month = month[-1]
    else:
        pass
    new_date = [month,
                date_splited[2],
                date_splited[0], 
                ]
    fixed_date = '-'.join(new_date)
    print(fixed_date)
    #return datetime.strptime(fixed_date, '%m-%d-%Y')
    
    year = int(date_splited[0])
    month = int(month)
    day = int(date_splited[2])
    if (fixed_date == "2-31-2019"):
        day = day -  3 
        print(f"new day {day}")
    elif (fixed_date == "9-31-2018"):
        day = day - 2
    elif (fixed_date == "2-29-2018"):
        day = day - 1
    elif (fixed_date == "2-29-2019"):
        day = day- 1
    elif (fixed_date == "2-30-2018"):
        day = day -3
    elif fixed_date == "11-31-2018":
        day = day -1
    elif fixed_date == "4-31-2019":
        day -=1
    elif fixed_date == "4-31-2018":
        day -=1
    elif fixed_date == "2-30-2019":
        day -=3
    return datetime(year,month ,day)

df_new["datestamp"] = df_new["date"].apply(lambda x: parse_date(x))
df_new.head()
```
