```python
import pandas as pd
import re
import unicodedata
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.text import Tokenizer

%matplotlib inline
```

```python
%ls
```

```python
df = pd.read_csv("clean_data/all_data.csv")
df = df.drop("Unnamed: 4", axis=1)
df.head()
```

```python
print(df["date"].unique())
```

```python
#df['numbers'] = df['document'].apply(lambda x: re.sub("[0-9]+", '', x))
def find_telefon_numbers(x):
    numbers = []
    result = re.findall("[0-9]+", x)
    for n in result:
        if (len(n)> 5):
            numbers.append(n)
    return numbers
         

def remove_accent(text):
    output = unicodedata.normalize('NFD', text).encode('ascii', 'ignore')
    decoded = output.decode('utf-8')
    return decoded

def remove_punctuatios(text):
    output = text.replace("")

df['numbers'] = df['document'].apply(lambda x: find_telefon_numbers(x))
df['lower_case'] = df['document'].str.lower()
df['lower_case'] = df['lower_case'].apply(lambda x: remove_accent(x))
df = df.sample(frac=1).reset_index(drop=True)

#lambda x: True if x % 2 == 0 else False

df.head()
```

```python
#df_demo = df.head(n=20)
#df_demo.to_json("demo_json.json")
#df_demo.to_csv("demo_data.csv", index = None, header=True)
```

```python
tokenizer = Tokenizer()
```

```python
import itertools

numbers = df["numbers"].tolist()
numbers = list(itertools.chain.from_iterable(numbers))

print(len(list(numbers)))
print(numbers[:5])
print(len(set(numbers)))
```

```python
tokenizer.fit_on_texts(numbers)

total_words = len(tokenizer.word_index) + 1
print(total_words)
```

```python
word_counts = dict(tokenizer.word_counts)
print(word_counts)
```

```python
word_counts["5212310"]
```

```python
df_counts = pd.DataFrame({'Value': list(word_counts.values()), 'Label': list(word_counts.keys()),
                         'x': np.arange(len(word_counts))}, index = list(word_counts.keys()) )


df_counts = df_counts.sort_values(['Value'], ascending=False)
df_counts.head(n=10)
```

```python
df_counts.shape
```

```python
def plot_word_count(df, title):
    plt.figure(figsize=(10,6))
    ax = sns.barplot(x="x", y="Value", data=df, order=df['x'])
    ax.set_xlabel(title)
    ax.set_ylabel('Number of Occurrences')
    ax.set_xticklabels(df['Label'], rotation='vertical', fontsize=10)
    
```

```python
top20_sorted = df_counts.iloc[0:40]
top20 = df_counts.iloc[0:40]

plot_word_count(top20_sorted, f"Top {len(top20_sorted)} Words")
```

```python
plt.figure(figsize=(16, 6))

sns.boxplot(x=df_counts['Value'])
```

```python
x_o = df_counts.loc[(df_counts["Value"] < 50) & (df_counts["Value"] > 1 )]["Value"].tolist()

plt.figure(figsize=(16, 6))
sns.distplot(x_o, rug=True,norm_hist=False, kde = True)
plt.show()
```

```python
# TIME SERIES for OutLayers

#df_new = df['document'].apply(lambda x: find_telefon_numbers(x))

def find_outlayer(numbers, target):
    status = False
    for x in numbers:
        if x == target:
            status=True
            break
    return status
            


df_new = df.loc[df['numbers'].apply(lambda x: find_outlayer(x,  "72457571")) ]
```

```python
df_new.head()
```

```python
def extract_list(my_list):
    value = ""
    for x in my_list:
        value = x
        break
    return value

df_new["numbers"] = df_new["numbers"].apply(lambda x: extract_list(x))
df_new.head()
```

```python
from datetime import timedelta, datetime
def parse_date(my_date):
    date_splited = my_date.split("-")
    month = date_splited[1]
    if (month[0] == "0"):
        month = month[-1]
    else:
        pass
    new_date = [month,
                date_splited[2],
                date_splited[0], 
                ]
    fixed_date = '-'.join(new_date)
    print(fixed_date)
    #return datetime.strptime(fixed_date, '%m-%d-%Y')
    
    year = int(date_splited[0])
    month = int(month)
    day = int(date_splited[2])
    if (fixed_date == "2-31-2019"):
        day = day -  3 
        print(f"new day {day}")
    elif (fixed_date == "9-31-2018"):
        day = day - 2
    elif (fixed_date == "2-29-2018"):
        day = day - 1
    elif (fixed_date == "2-29-2019"):
        day = day- 1
    elif (fixed_date == "2-30-2018"):
        day = day -3
    elif fixed_date == "11-31-2018":
        day = day -1
    elif fixed_date == "4-31-2019":
        day -=1
    elif fixed_date == "4-31-2018":
        day -=1
    elif fixed_date == "2-30-2019":
        day -=3
    return datetime(year,month ,day)

df_new["datestamp"] = df_new["date"].apply(lambda x: parse_date(x))
df_new.head()
```

```python
grouped = df_new.groupby('datestamp').count()
```

```python
grouped.head()
```

```python
ninety_five_quant = grouped.quantile(.95)[0] # [0] since we only need the quantile value
ninety_ninth_quant = grouped.quantile(.99)[0] # [0] since we only need the quantile value
ninety_eight_quant = grouped.quantile(.98)[0] # [0] since we only need the quantile valueninety_five_quant = grouped.quantile(.95)[0] # [0] since we only need the quantile value
ninety_ninth_quant = grouped.quantile(.99)[0] # [0] since we only need the quantile value
ninety_eight_quant = grouped.quantile(.98)[0] # [0] since we only need the quantile value
```

```python
count = df_new["date"].size

total_req =  'Total request: ' + str(count)
nine_five_quant_str =  '95th Quantile: ' + str(ninety_five_quant)
nine_eight_quant_str =  '98th Quantile: ' + str(ninety_eight_quant)
nine_nine_quant_str = '99th Quantile: ' + str(ninety_ninth_quant)

# x and y axis values are extracted from the grouped DataFrame
x = grouped.index
y = grouped.values

print ('Plotting Graph..')
fig = plt.figure(figsize=(16, 6))
fig.suptitle('Scatter Plot', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.95)

ax.set_xlabel('Request Time')
ax.set_ylabel('Request Count')

text_x_axis_value = 0.85
ax.text(text_x_axis_value, 0.96, total_req, horizontalalignment='center', verticalalignment='center', transform = ax.transAxes)
ax.text(text_x_axis_value, 0.88, nine_five_quant_str, horizontalalignment='center', verticalalignment='center', transform = ax.transAxes)
ax.text(text_x_axis_value, 0.80, nine_eight_quant_str, horizontalalignment='center', verticalalignment='center', transform = ax.transAxes)
ax.text(text_x_axis_value, 0.74, nine_nine_quant_str, horizontalalignment='center', verticalalignment='center', transform = ax.transAxes)
ax.plot_date(x, y, xdate=True, ydate=False, color='skyblue')
plt.show()
```
